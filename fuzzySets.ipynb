1. Let A(x) and B(x) be two fuzzy sets of speed limiton a highway, defined by the following
membership functions:
a. Fuzzy set A: “Low speed limit”

μ
A(x) = {1, if x < 30 (low speed)
x−30
50−30
, if 30 < x < 50 (moderately low speed)
x−50
70−50
, if 50 < x < 70 (increasing speed)
0, if x > 70 (not low speed)}

b. Fuzzy set B: “High speed limit”

μ
B(x) = {0, if x < 60 (Not high speed)
x−60
80−60
, if 60 < x < 80 (moderately high speed)
x−80
100−80
, if 80 < x < 100 (increasing speed)
1, if x > 100 (high speed)}

Now, perform the following operations to obtain the resultant set and plot the same:




i. Union - (A v B)x












ii. Intersection - (A ∧ B)x










iii. Complement of both sets - A











iv. Difference of the two sets

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fuzzy_set_A(x):
    if x < 30:
        return 1
    elif 30 <= x < 50:
        return (x - 30) / (50 - 30)
    elif 50 <= x < 70:
        return (70 - x) / (70 - 50)
    else:
        return 0
```

```python
def fuzzy_set_B(x):
    if x < 60:
        return 0
    elif 60 <= x < 80:
        return (x - 60) / (80 - 60)
    elif 80 <= x < 100:
        return (100 - x) / (100 - 80)
    else:
        return 1
```

```python
x_values = np.linspace(0, 120, 500)
A = np.array([fuzzy_set_A(x) for x in x_values])
B = np.array([fuzzy_set_B(x) for x in x_values])
```

```python
# Union
Union = np.maximum(A, B)
```

```python
# Intersection
Intersection = np.minimum(A, B)
```

```python
# Complement
Complement_A = 1 - A
Complement_B = 1 - B
```

```python
# Difference
Difference = np.maximum(A - B, 0)  # Ensure no negative values
```

```python
plt.figure(figsize=(10, 10))
plt.subplot(2, 2, 1)
plt.plot(x_values, A, label='A(x)')
plt.plot(x_values, B, label='B(x)')
plt.title('Membership Functions A and B')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 2)
plt.plot(x_values, Union, label='A ∪ B (Union)')
plt.title('Union of A and B')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
```

```python
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 3)
plt.plot(x_values, Intersection, label='A ∩ B (Int)')
plt.title('Intersection of A and B')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
```

```python
plt.figure(figsize=(13, 13))

plt.subplot(2, 2, 4)
plt.plot(x_values, Complement_A, label='A̅ (Comp of A)')
plt.plot(x_values, Complement_B, label='B̅ (Comp of B)')
plt.title('Complement of A and B')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
```

```python
# Plot Difference separately
plt.figure(figsize=(5, 5))
plt.plot(x_values, Difference, label='A - B (Difference)')
plt.title('Difference of A and B')
plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.legend()
plt.show()
```

2. Let C(x) and D(x) be two fuzzy sets of temperature (Celsius) in a city, defined by the
following membership functions:

a. Fuzzy set C: “Cold”
μ
C(x) = {0, if x <− 10 (extremely cold)
x−(−10)
0−(−10)
, if − 10 < x < 0(very cold)
1, if0 < x < 5(cold)
0, if x > 5 (not cold)}

b. Fuzzy set D: “Warm”
μ
D(x) = {0, if x < 25 (Not warm)
x−25
35−25
, if 25 < x < 35(moderately warm)
1, if 35 < x < 40(warm)
0, if x > 40 (extremely warm)}

Now, perform the following operations to obtain the result:













i. Max-Min Composition

ii. Max-Product Composition


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def fuzzy_set_C(x):
    if x < -10:
        return 0
    elif -10 <= x < 0:
        return (x - (-10)) / (0 - (-10))
    elif 0 <= x < 5:
        return 1
    else:
        return 0
```

```python
def fuzzy_set_D(x):
    if x < 25:
        return 0
    elif 25 <= x < 35:
        return (x - 25) / (35 - 25)
    elif 35 <= x < 40:
        return 1
    else:
        return 0
```

```python
# Define the range of temperatures
x_values_temp = np.linspace(-20, 50, 500)
```

```python
# Compute membership values for C and D
C_values = np.array([fuzzy_set_C(x) for x in x_values_temp])
D_values = np.array([fuzzy_set_D(x) for x in x_values_temp])
```

```python
# Compute Max-Min Composition
def max_min_composition(C_values, D_values, x_values_temp):
    result = np.zeros_like(D_values)
    for i in range(len(D_values)):
        result[i] = np.max(np.minimum(C_values, D_values[i]))
    return result
```

```python
# Compute Max-Product Composition
def max_product_composition(C_values, D_values):
    result = np.zeros_like(D_values)
    for i in range(len(D_values)):
        result[i] = np.max(C_values * D_values[i])
    return result
```

```python
# Calculate compositions
Max_Min_Composition = max_min_composition(C_values, D_values, x_values_temp)
Max_Product_Composition = max_product_composition(C_values, D_values)
```

```python
# Plotting
plt.figure(figsize=(10, 6))
```

```python
# Max-Min Composition Plot
plt.subplot(1, 2, 1)
plt.plot(x_values_temp, Max_Min_Composition, label='Max-Min Composition')
plt.title('Max-Min Composition of C and D')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)
```

```python
# Max-Product Composition Plot
plt.subplot(1, 2, 2)
plt.plot(x_values_temp, Max_Product_Composition, label='Max-Product Composition')
plt.title('Max-Product Composition of C and D')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Membership Degree')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
